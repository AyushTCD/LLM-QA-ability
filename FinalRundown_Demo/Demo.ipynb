```python
import time
from llama_index.llms.ollama import Ollama
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer
from sacrebleu import corpus_ter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Initialization of the LLM
model_choice = "llama2"  # Change as necessary
llm = Ollama(model=model_choice, request_timeout=30.0)

# Asking the LLM a question
question = "Tell me about the battle of Boyne in 100 words"

start_time = time.time()
response = llm.complete(question)
end_time = time.time()

# Assuming these are your reference responses
reference_responses = [
    "The Battle of Boyne, occurring in 1690, was a pivotal event in Irish history, marking a decisive victory for William III over James II.",
    "In 1690, the Battle of Boyne saw William III defeat James II, significantly impacting the political landscape of Ireland."
]

# Displaying the LLM response and time taken
print(response)
print(f"Time taken: {end_time - start_time} seconds")

# Proceed with NLP metrics evaluation
# Assuming `response` is a string containing the LLM-generated text
llm_response = response['completions'][0]['completion']  # Adjust based on the actual structure of `response`

# BLEU score calculation
bleu_score = sentence_bleu([ref.split() for ref in reference_responses], llm_response.split())
print(f"BLEU Score: {bleu_score}")

# ROUGE score calculation
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_score = scorer.score(reference_responses[0], llm_response)  # Using the first reference for simplicity
print(f"ROUGE Scores: {rouge_score}")

# TER score calculation (using sacreBLEU's TER as an approximation)
ter_score = corpus_ter([llm_response], [[ref] for ref in reference_responses]).score
print(f"TER Score: {ter_score}")

# Cosine similarity calculation (using TF-IDF)
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform([llm_response] + reference_responses)
cosine_similarities = cosine_similarity(tfidf[0:1], tfidf[1:])
print(f"Cosine Similarities: {cosine_similarities.flatten()}")
```
